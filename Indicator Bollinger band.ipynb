# Bollinger band On Historical data of SUNPHARMA 

from (2018-01-29 to 2019-04-26 )

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
a = pd.read_csv("SUNPHARMA2.csv",usecols=["Date","Open","High","Low","Close"])
def BollingerBand(df, column="Close", period=20):
    sma = df[column].rolling(window=period, min_periods=period - 1).mean()
    std = df[column].rolling(window=period, min_periods=period - 1).std()
    up = (sma + (std * 2)).to_frame('BBANDUP')
    lower = (sma - (std * 2)).to_frame('BBANDLO')
    return df.join(up).join(lower)

plt.style.use('seaborn')
c=BollingerBand(a)
c["Close"].plot()
c['BBANDLO'].plot()
c["BBANDUP"].plot()
plt.show()
print (" \t \t' ABOVE GRAPH OF  Bollinger band  ' \n \n ")
c['BBANDLO'].plot()
c["BBANDUP"].plot()
plt.show()
print (" \t ' ABOVE GRAPH OF  Bollinger band UP & Low ' \n \n ")
```
